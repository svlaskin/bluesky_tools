This is a tutorial for the EnvImport module. This can be used for several applications, such as:

1. Importing building and airfield geofence geometry for a certain cutoff height for a given city/bounding-box.
2. Importing all delviery locaitons for a given city/bounding-box
3. Importing hospital locations

Let us import the module first, after which we can import the data for a given city (here, we will use Amsterdam).


```python
from envimport import *
data = envimport("Amsterdam",50)
```

Now run the module. The city name string and ctoff height for building geofences are user-defined variables.
A Lat/Lon bounding box can also be used in place of the city name.
The outputs are hospital locations, geofence geometry and delivery locations

```python
name = "Amsterdam"
cutoff = 50
hopsitals, geofences, delivery_locations = envimport(name, cutoff)
```

```python
# Let's print one geofence:
print(geofences[0])
geofences[15][1]
```

For the GEOFENCE plugin, this is now quite simple to import.
In the 'geofences' variable have a list of tuples containing the name and coordinates such that they can be imported directly via stack as follows:

```python
stack.stack("PLUGIN IMPORT GEOFENCE")
for geofence in geofences:
    geofence_name = geofence[0]
    geofence_coordinates = geofence[1]
    stack.stack("GEOFENCE {geofence_name} {9000} {0} {coordinates}")

stack.stack("drawgeofence")
```

If you wish to only import building geofences without making a plugin, then you can make use of the GEOFENCE plugin as follows:

```python
PLUGIN IMPORT GEOFENCE 
IMPORTGEOFENCES cutoff_height, input
```

where input is either the city name (ie Amsterdam) or the bounding box, in the form Lat_min Lon_min Lat_max Lon_max.
